### 1. What is name of car that has the highest horsepower?  

```python
#  Find and display the highest horsepower car with its stats
max_hp_row = df[df['horsepower'] == df['horsepower'].max()]
car_name_max_hp = max_hp_row['name'].values[0]
car_hp_max = max_hp_row['horsepower'].values[0]

print(f"\nThe highest horsepower car is: '{car_name_max_hp}'")
print(f"Its horsepower is: {car_hp_max} hp")
print("\nFull stats of this car:")
print(max_hp_row)
```

### 2. How many cars have mpg ≥ 35?

```python
# Count cars with mpg >= 35
high_mpg_cars = df[df['mpg'] >= 35]
num_high_mpg = len(high_mpg_cars)

print(f"Number of cars with MPG ≥ 35: {num_high_mpg}")
print("\nThese high-MPG cars:")
print(high_mpg_cars[['name', 'mpg']])  # Showing name and mpg for verification
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
# Filter cars meeting both conditions
filtered_cars = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]

# Count occurrences of each origin
origin_counts = filtered_cars['origin'].value_counts()

# Get the most common origin
most_common_origin = origin_counts.idxmax()
count = origin_counts.max()

print(f"The most common origin for cars with horsepower > 100 and weight < 3000 is: {most_common_origin}")
print(f"Number of such cars: {count}")
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
# Filter Japanese cars and calculate mean acceleration
japanese_cars = df[df['origin'] == 'japan']
mean_accel_japan = round(japanese_cars['acceleration'].mean(), 2)

print(f"Mean acceleration of Japanese cars: {mean_accel_japan} seconds (0-60 mph)")
```

### 5. Which year had the highest average mpg?

```python
# Calculate average MPG by year
avg_mpg_by_year = df.groupby('model_year')['mpg'].mean()

# Find the year with highest average MPG
best_year = avg_mpg_by_year.idxmax()
highest_avg_mpg = round(avg_mpg_by_year.max(), 2)

print(f"The year with highest average MPG is: {best_year}")
print(f"Average MPG that year: {highest_avg_mpg}")
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better visuals
sns.set(style="whitegrid")

# 1. Calculate average MPG by year and origin
avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# 2. Create the plot
plt.figure(figsize=(12, 6))

# Using Seaborn's lineplot (automatically handles colors and legends)
sns.lineplot(data=avg_mpg, 
             x='model_year', 
             y='mpg', 
             hue='origin',
             style='origin',
             markers=True,
             dashes=False,
             linewidth=2.5,
             markersize=8)

# 3. Customize the plot
plt.title('Evolution of Average MPG by Origin (1970-1982)', fontsize=14, pad=20)
plt.xlabel('Model Year', fontsize=12)
plt.ylabel('Average MPG', fontsize=12)
plt.xticks(avg_mpg['model_year'].unique())  # Show all years on x-axis

# Add grid and adjust legend
plt.grid(alpha=0.3)
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show plot
plt.tight_layout()
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
import pandas as pd

# 1. Filter cars that appear in multiple years
multi_year_cars = df.groupby('name').filter(lambda x: len(x['model_year'].unique()) > 1)

# 2. Calculate stats for each model
consistency_stats = (multi_year_cars.groupby('name')
                     .agg(
                         appearances=('model_year', 'count'),
                         mpg_std=('mpg', 'std'),
                         avg_mpg=('mpg', 'mean')
                     )
                     .reset_index())

# 3. Filter for consistent models (std < 1.0 and at least 2 appearances)
consistent_cars = consistency_stats[
    (consistency_stats['mpg_std'] < 1.0) & 
    (consistency_stats['appearances'] >= 2)
]

# 4. Sort and get the most consistent models
result = (consistent_cars
          .sort_values(['appearances', 'avg_mpg'], ascending=[False, False])
          .reset_index(drop=True))

# 5. Format and display results
if not result.empty:
    print("Most Consistent Car Models (by MPG):\n")
    print(result[['name', 'appearances', 'avg_mpg', 'mpg_std']]
          .rename(columns={
              'name': 'Model',
              'appearances': 'Years Produced',
              'avg_mpg': 'Average MPG',
              'mpg_std': 'MPG Std Dev'
          })
          .to_string(index=False))
else:
    print("No cars met the consistency criteria (std(mpg) < 1.0 across multiple years)")
```
